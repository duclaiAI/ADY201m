# Data Collection, Understanding, Preparation


**Date**: Ngày giao dịch của cổ phiếu.

**Open**: Giá mở cửa – giá của cổ phiếu tại thời điểm bắt đầu phiên giao dịch trong ngày.

**High**: Giá cao nhất mà cổ phiếu đạt được trong phiên giao dịch của ngày.

**Low**: Giá thấp nhất mà cổ phiếu đạt được trong phiên giao dịch của ngày.

**Close**: Giá đóng cửa – giá của cổ phiếu tại thời điểm kết thúc phiên giao dịch trong ngày.

**Volume**: Khối lượng giao dịch – tổng số cổ phiếu được giao dịch trong ngày (bao gồm mua và bán).

**Name**: Tên của công ty hoặc mã cổ phiếu (ticker symbol).

### Feature Engineering

```python
# Drop Name and Date Columns because we wouldn't use it to train model.
stock_data = stock_data[['open', 'high', 'low', 'close']]
stock_data.head(5)
```

# Regression Analysis

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
X = stock_data[['open', 'high', 'low']]
Y = stock_data['close']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression().fit(X_train, Y_train)
predictions = model.predict(X_test)
```

```python
plt.scatter(predictions, Y_test, linewidths=6)
plt.plot([Y.min(), Y.max()], [Y.min(), Y.max()], 'k--', lw=2, label='Ideal Line')
plt.title("Linear Regression Model's Performance")
plt.xlabel('Predicted Stock')
plt.ylabel('Actual Stock')
plt.legend();
```

### Performance

```python
import sklearn.metrics as sm
print("Mean Squared Error:", round(sm.mean_squared_error(Y_test, predictions),5))
print("Median absolute error =", round(sm.median_absolute_error(Y_test, predictions), 2))
print("Mean absolute error =", round(sm.mean_absolute_error(Y_test, predictions), 2)) 
```
